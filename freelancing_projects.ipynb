प्रोजेक्ट 1: CSV डेटा से Department-wise Average Salary निकालना

```python
import csv #csv file raed
from collections import defaultdict 
```

```python
with open('employees.csv','w',newline='')as file:
    writer= csv.writer(file)
    writer.writerow(['ajay','Department','salary'])
    writer.writerow(['ajay','IT',50000])
    writer.writerow(['sunil','HR',40000])
    writer.writerow(['reeta','IT',55000])
    writer.writerow(['neelam','Finance',60000])
    writer.writerow(['suresh','HR',45000])
```

```python
department_salary=defaultdict(list)
with open('employees.csv','r')as file:
    reader=csv.DictReader(file)
    for row in reader:
        department_salary[row['Department']].append(int(row['salary']))
        department_salary
        
```

```python
department_avg={dept:sum(salaries)/len(salaries)
                for dept,salaries in department_salary.items()}
department_avg
```

```python
print("Department-wise Average Salary:")
for dept,avg in department_avg.items():
    print(f"{dept}:{avg}")
```

प्रोजेक्ट 2: Text File से Word Frequency निकालना

```python
text="""python a popular programming language hai|python ka use web development ,data science , aur automation me hota hai|"""
with open('article.txt','w',encoding='utf-8') as f:
    f.write(text)
```

```python
with open('article.txt','r',encoding='utf-8')as f:
    content=f.read()
    content
```

Cell 3: Word list बनाना

```python
import re 
words =re.findall(r'\b\w+\b',content)
words
```

Cell 4: Word frequency निकालना

```python
from collections import Counter
word_count=Counter(words)
word_count
```

```python
print("word Frequency:")
for word,count in word_count.items():
    print(f"{word}:{count}")
```
